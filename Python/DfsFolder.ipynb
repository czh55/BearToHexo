```python
import os
import re

class Article:

    def __init__(self, name):
        self.name = name
        self.dir = name.strip('.md')
        # print(self.name)
        # print(self.dir)

    def run(self):
        # 获取文件名filename
        strlist = self.name.split('/') # 用逗号分割str字符串，并保存到列表
        #print(len(strlist))
        #print(strlist[5])
        filename = strlist[5]
        
        # 在复制的文件中修改，并删除旧的
        f = open(self.name, 'r', encoding='UTF-8')
        f_new = open(self.name + '.new', 'w', encoding='UTF-8')
        print('正在修改' + self.name + '的图片路径')
        lines = f.readlines()
        for line in lines:
            picture_path = re.findall(r'!\[.{0,100}\]\(.{0,10000}\)', line)
            # print(line)
            if picture_path:
                picture_path = picture_path[0]
                # 保证编码正确
                picture_path.encode('utf-8')
                filename.encode('utf-8')
                # 旧的编码：
                print("旧的编码："+picture_path)
                print(picture_path.rfind('/'))
                new_path = picture_path[0 : picture_path.index('(') + 1] + picture_path[picture_path.rindex('/'): ] + '\r\n'
                # 新的编码：
                print("新的编码："+new_path)
                f_new.write(new_path)
            else:
                f_new.write(line)
        f.close()
        f_new.close()
        # 换文件
        os.remove(self.name)
        os.rename(self.name + '.new', self.name)
        print('修改完毕')
        
        
def dfsFloder(path):
    article_list = []
    dir_list = []
    dir_list_dfs = []
    article_with_picture_list = []
    files = os.listdir(path)
    # print(files)
    
    # 找到当前文件夹的下的文件和文件夹
    for f in files:
        if '.md' in f:
            article_list.append(f)
        elif '.' not in f:
            dir_list.append(f)
    #print(article_list)
    #print(dir_list)         
    
    # 查询到本目录下有以下文章的图片路径需要修改+子目录
    for dir in dir_list:
        flag = 0;
        for article in article_list:
            if dir in article:
                flag = 1
                article_with_picture_list.append(path+'/'+article)
        if flag == 0:
             dir_list_dfs.append(path+'/'+dir)
    print('查询到本目录下有以下文章的图片路径需要修改：')
    for article_with_picture in article_with_picture_list:
        print(article_with_picture)
    print('查询到本目录有子目录：')
    for dir_dfs in dir_list_dfs:
        print(dir_dfs)
        
    # 当前目录执行转换
    for article_with_picture in article_with_picture_list:
        article = Article(article_with_picture)
        article.run()
        
    # 遍历子目录
    for dir_dfs in dir_list_dfs:
        dfsFloder(dir_dfs)
    
    
    

def main():
    dfsFloder('/Users/czh/BearTemp/&&&')

    print('已完成完成！')

if __name__ == "__main__":
    main()
```
